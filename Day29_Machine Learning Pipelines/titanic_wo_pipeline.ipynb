```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
df.head(3)
```

```python
df = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
df.head(3)
```

```python
# train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']), df['Survived'], test_size=0.2)
X_train.head(3)
```

```python
df.isnull().sum()
```

```python
# Simple Imputer -> Age, Embarked
si_age = SimpleImputer(strategy='median')
si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age_imp = si_age.fit_transform(X_train[['Age']])
X_train_embarked_imp = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age_imp = si_age.transform(X_test[['Age']])
X_test_embarked_imp = si_embarked.transform(X_test[['Embarked']])

X_train_age_imp
```

```python
df.isnull().sum()
```

```python
X_test_embarked
```

```python
# One Hot Encoding -> Sex, Embarked
ohe_sex = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked_imp)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked_imp)

X_train_embarked
```

```python
X_train.head(2)
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])

X_train_rem
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age_imp, X_train_sex, X_train_embarked), axis =1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age_imp, X_test_sex, X_test_embarked), axis =1)

X_train_transformed.shape
```

```python
X_train_transformed
```

```python
# Model Training
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed, y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```
