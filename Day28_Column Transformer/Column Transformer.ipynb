```python
import pandas as pd
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['has_covid']), df['has_covid'], test_size=0.2)
```

```python
X_train
```

## 1. Aam Zindagi

```python
# Simple Imputer -> Fever
si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[['fever']])

# Also the test data
X_test_fever = si.transform(X_test[['fever']])

X_train_fever.shape
X_test_fever.shape
```

```python
# Ordinal Encoder -> Cough
oe = OrdinalEncoder(categories=[['Mild', 'Strong']])
X_train_cough = oe.fit_transform(X_train[['cough']])
X_test_cough = oe.transform(X_test[['cough']])
X_train_cough.shape
X_test_cough.shape
```

```python
# One Hot Encoder -> gender, city
ohe = OneHotEncoder(drop='first', sparse_output=False)
X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])
X_test_gender_city = ohe.transform(X_test[['gender', 'city']])
X_train_gender_city.shape
X_test_gender_city.shape
```

```python
# Extracting Age
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values
X_train_age.shape
```

```python
# Combine all transformed columns
X_train_transformed = np.concatenate((X_train_age, X_train_fever, X_train_gender_city, X_train_cough),axis=1)
X_train_transformed.shape
```

```python
X_test_transformed = np.concatenate((X_test_age,X_test_fever,X_test_gender_city,X_test_cough),axis=1)
X_test_transformed.shape
```

```python
X_train_transformed=np.round(pd.DataFrame(X_train_transformed),0)
X_train_transformed
```

```python
X_test_transformed=np.round(pd.DataFrame(X_test_transformed),0)
X_test_transformed
```

## 2. Mentos Zindagi

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(
    transformers=[
        ('ct1', SimpleImputer(), ['fever']),
        ('ct2', OrdinalEncoder(categories=[['Mild','Strong']]), ['cough']),
        ('ct3', OneHotEncoder(sparse_output=False, drop='first'), ['gender','city'])
    ],
    remainder='passthrough'
)
```
