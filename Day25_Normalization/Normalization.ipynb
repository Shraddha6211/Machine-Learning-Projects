```python
# Import Dependencies
import numpy as np #Linear Algebra
import pandas as pd #Data Processing
import matplotlib.pyplot as plt # Data Visualization
import seaborn as sns
```

```python
# Read dataset
df = pd.read_csv("wine_data.csv", usecols = [0,1,2], header = None)
df.columns = ['Wine Type', 'Alcohol', 'Malic Acid']
df.sample(5)
```

```python
df.info()
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic Acid'])
```

```python
color_dict={1:'red',
           2:'blue',
           3:'green'}
sns.scatterplot(
    x='Alcohol',
    y='Malic Acid',
    hue='Wine Type',
    data=df,
    palette=color_dict
)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Wine Type', axis =1), df['Wine Type'], test_size = 0.3, random_state = 0)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
#fit the scaler to the training dataset, it will learn the parameters
X_train_scaled = scaler.fit_transform(X_train)

#transform train and test sets
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
X_train_scaled
```

```python
# convert this numpy array to pandas dataframe
X_train_scaled = pd.DataFrame(X_train_scaled, columns= X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
X_test_scaled.head(5)
```

```python
X_train_scaled.head(5)
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
# Side by Side Comparision
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize= (12,5))

ax1.scatter(X_train['Alcohol'], X_train['Malic Acid'],c=y_train)
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Alcohol'], X_train_scaled['Malic Acid'],c=y_train)
ax2.set_title("After Scaling")
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize =(12,5))

# Before Scaling
ax1.set_title("Before Scaling")
sns.kdeplot(X_train['Alcohol'], ax= ax1, label='Alcohol')
sns.kdeplot(X_train['Malic Acid'], ax = ax1, label='Malic Acid')
ax1.legend()

# After Scaling
ax2.set_title("After Scaling")
sns.kdeplot(X_train_scaled['Alcohol'], ax= ax2, label='Alcohol')
sns.kdeplot(X_train_scaled['Malic Acid'], ax = ax2, label='Malic Acid')
ax2.legend()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize= (12,5))

#Before Scaling
ax1.set_title("Alcohol Distribution Before Scaling")
sns.kdeplot(X_train['Alcohol'], ax=ax1)
# After Scaling
ax2.set_title("Alcohol Distribution After Scaling")
sns.kdeplot(X_train_scaled['Alcohol'],ax =ax2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

# Before Scaling
ax1.set_title("Malic Acid Distribution Before Scaling")
sns.kdeplot(X_test['Malic Acid'], ax=ax1, label="Malic Acid")
ax1.legend()
# After Scaling
ax2.set_title("Malic Acid Distribution After Scaling")
sns.kdeplot(X_test['Malic Acid'], ax=ax2, label="Malic Acid")
ax2.legend()
```
