Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis
PIMA Diabetes Dataset

```python
diabetes_dataset= pd.read_csv('/content/Datasets/diabetes.csv')
print(diabetes_dataset.head())
```

```python
print(diabetes_dataset.describe)
```

```python
print(diabetes_dataset)
```

```python
diabetes_dataset['Outcome'].value_counts()
```

0 --> Non-Diabetic
1 --> Diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
```

Separating Data and Labels


```python
X= diabetes_dataset.drop(columns='Outcome', axis=1)
Y= diabetes_dataset['Outcome']
```

```python
print(X.head())
print(Y.head())
```

Data Standarization

Training and Test Data Split

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X_train.shape, X_test.shape)
```

Model Training

```python
classifier=svm.SVC(kernel='linear')
```

```python
#training the support vector machine Classifier
classifier.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on the training Data
X_train_prediction= classifier.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
print('Accuracy score of the training data: ', training_data_accuracy*100)
```

```python
# accuracy score on the Testing Data
X_test_prediction= classifier.predict(X_test)
testing_data_accuracy= accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the testing data: ', testing_data_accuracy*100)
```

Making a Predictive System

```python
#input_data= input('Enter the values as : Pregnancies	Glucose	BloodPressure	SkinThickness	Insulin	BMI	DiabetesPedigreeFunction	Age')
#print('You entered: ', input_data)
input_data=(0,118,84,47,230,45.8,0.551,31)

#Change this input data to numpy array
input_data_as_numpy_array= np.asarray(input_data)

#Reshape the array as we are predicting for one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

prediction= classifier.predict(input_data_reshaped)
print(prediction)
```
